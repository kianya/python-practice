## С использованием t-test статистики проверить взаимосвязи между несколькими переменными

-----

```python
import scipy as sc
```

Проверим взаимосвязи между распределениями. Перед проверкой надо их нормализовать.  
Сначала проверим взаимосвязь между horsepower и engine-size

```python
sc.stats.ttest_ind((df['horsepower']-df['horsepower'].max())/df['horsepower'].max().astype(np.float64),(df['engine-size']-df['engine-size'].max())/df['engine-size'].max().astype(np.float64),equal_var = False)
```

Мы не можем отвергнуть нулевую гипотезу.   
Теперь проверим взаимосвязь между city-mpg и highway-mpg

```python
sc.stats.ttest_ind((df['city-mpg']-df['city-mpg'].mean())/df['city-mpg'].max().astype(np.float64),(df['highway-mpg']-df['highway-mpg'].mean())/df['highway-mpg'].max().astype(np.float64), equal_var = False)
```

Тут мы тоже не можем отвергнуть нулевую гипотезу, но наша уверенность в том, что есть взаимосвязь намного больше, нежели в предыдущем случае, поскольку p-value = 0.99
